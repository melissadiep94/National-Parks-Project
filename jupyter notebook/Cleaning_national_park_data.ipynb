```python
import os
import pandas as pd
import numpy
import urllib.request, json
import time
from datetime import date
import requests
import pymongo


#import API key
from config2 import APIKEY
```

```python
from natlparks import NatlParks
parks = NatlParks(APIKEY)
```

```python
parks_info= parks.parks(limit=500)
#parks_info
```

```python
#Transform into a dataframe with below info...

#print out names of all national parks and their parkCode
#park description
#long and latitude of each park
#address
#link to each park website
#hours of operation
#image of each park
#weatherInfo
#print out activities for each park

```

```python
#pd.DataFrame.from_dict(parks_info['data'])
```

```python
df = pd.DataFrame(parks_info['data'])

df
```

```python
# For every 'x' in the activities column, extract the values in the 'name' key and put them in a single list
# Iterates over the rows of the dataframe
df['activity_names'] = df['activities'].apply(lambda x: [k['name'] for k in x])
df['activity_ids'] = df['activities'].apply(lambda x: [k['id'] for k in x])
df['images_url'] = df['images'].apply(lambda x: [k['url'] for k in x])
df['topic_names'] = df['topics'].apply(lambda x: [k['name'] for k in x])
df['topic_ids'] = df['topics'].apply(lambda x: [k['id'] for k in x])
df['addresses_postalCode'] = df['addresses'].apply(lambda x: [k['postalCode'] for k in x])
df['addresses_city'] = df['addresses'].apply(lambda x: [k['city'] for k in x])
df['addresses_stateCode'] = df['addresses'].apply(lambda x: [k['stateCode'] for k in x])
df['addresses_street'] = df['addresses'].apply(lambda x: [k['line1'] for k in x])
df['standardHours'] = df['operatingHours'].apply(lambda x: [k['standardHours'] for k in x])
df['entranceFees_cost'] = df['entranceFees'].apply(lambda x: [k['cost'] for k in x])
df['entranceFees_description'] = df['entranceFees'].apply(lambda x: [k['description'] for k in x])
df.head(2)
```

```python
#Drop columns not needed

df_clean= df.drop(columns=['id', 'activities', 'topics', 'contacts', 'entranceFees', 'entrancePasses' , 'fees', 'operatingHours', 'addresses', 'images', 'name'])
```

```python
#Re-order columns

df_clean = df_clean.reindex(columns=
                                     ['parkCode',
                                     'fullName',
                                     'url',
                                     'designation',
                                     'description',
                                     'latitude',
                                     'longitude',
                                     'latLong',
                                     'states',
                                     'directionsInfo',
                                     'directionsUrl',
                                     'weatherInfo',
                                     'activity_names',
                                     'activity_ids',
                                     'images_url',
                                     'topic_names',
                                     'topic_ids',
                                     'addresses_postalCode',
                                     'addresses_city',
                                     'addresses_stateCode',
                                     'addresses_street',
                                     'standardHours',
                                     'entranceFees_cost',
                                     'entranceFees_description']  
                                     
                                     
                                     )

df_clean.head(2)
```
